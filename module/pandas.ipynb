## pandas合并

```python
df1 = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['a','b','c','d'])
df2 = pd.DataFrame(np.arange(12,24).reshape((3,4)),columns=['a','b','c','d'])
df3 = pd.DataFrame(np.arange(24,36).reshape((3,4)),columns=['a','b','c','d'])
print(df1,df2,df3)
```

```python
df4 = pd.concat([df1,df2,df3],axis=0)#纵向合并
df4
```

```python
df4 = pd.concat([df1,df2,df3],axis=1)#横向合并
df4
```

```python
df4 = pd.concat([df1,df2,df3],axis=0,ignore_index=True)#纵向合并，index重新排序
df4
```

```python
df4 = pd.concat([df1,df2,df3],axis=1,ignore_index=True)#h横向合并，columns重排
df4
```

```python
df1 = pd.DataFrame(np.arange(12,24).reshape((3,4)),columns=['a','c','e','f'])
df2 = pd.DataFrame(np.arange(24,36).reshape((3,4)),columns=['a','b','d','f'])
df3 =pd.concat([df1,df2],join='outer',ignore_index=True,sort=False)
df3
```

```python
join='inner'仅合并公共的columns
```

```python
df3 = pd.concat([df1,df2],join='inner',ignore_index=True,sort=False)
df3
```

```python
df1 = pd.DataFrame(np.arange(12,24).reshape((3,4)),columns=['a','c','e','f'])
df2 = pd.DataFrame(np.arange(24,36).reshape((4,3)),columns=['a','b','d'])
df3 =pd.concat([df1,df2],axis=1,join_axes=[df1.index])#横向合并时，只考虑df1的index
df3
```

## pandas的merge合并

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D2','D3','D4']})
print(left,right)                    
```

```python
res = pd.merge(left,right,on='key')
print(res)
```

```python
left = pd.DataFrame({'key1':['K0','K0','K2','K3'],
                     'key2':['K0','K1','K0','K1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                      'key2':['K0','K0','K0','K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D2','D3','D4']})
print(left)
print(right)
```

```python
res = pd.merge(left,right,on=['key1','key2'],how='outer')#默认为inner
res
```

```python
res = pd.merge(left,right,on=['key1','key2'],how='inner')#只合并有相同key1和key2的
res
```

```python
res = pd.merge(left,right,on=['key1','key2'],how='left')#看得出来这是第一种outer方式截取前面四行
res
```

```python
res = pd.merge(left,right,on=['key1','key2'],how='right')#只考虑右边的
res
```

```python
res = pd.merge(left,right,on=['key1','key2'],how='outer',indicator=True)#默认为inner
res
```

## 合并两张index值不一样，且columns不一样的表

```python
left = pd.DataFrame({ 'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']},
                     index=['K0','K1','K2','K3'])
right = pd.DataFrame({'C':['C0','C1','C2','C3'],
                      'D':['D0','D2','D3','D4']},
                      index=['K0','K1','K4','K5'])
print(left)
print(right)
```

```python
res = pd.merge(left,right,left_index=True,right_index=True,how='outer')
res
```

```python
boys = pd.DataFrame({'k':['k0','k1','k2'],
                    'age':[1,2,3]})
girls = pd.DataFrame({'k':['k0','k0','k3'],
                     'age':[4,5,6]})
print(boys)
print(girls)
```

```python
合并的两张表中，有相同的columns值
```
